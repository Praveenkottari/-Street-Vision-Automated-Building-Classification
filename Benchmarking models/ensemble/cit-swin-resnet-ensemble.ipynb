```python
import torch
import torchvision.models as models
import torchvision.transforms as transforms
from sklearn import svm
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

from torch.utils.data import DataLoader, random_split
from torchvision import datasets, transforms

from sklearn.metrics import confusion_matrix, classification_report
```

```python
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
```

```python
def extract_features(loader):
    #model.eval()  # Set the model to evaluation mode
    features = []
    labels = []

    with torch.no_grad():  # Disable gradient calculation for efficiency
        for images, label in loader:
            images = images.to(device)  # Move images to the same device as the model
            image_features = resnet_model(images).view(images.size(0), -1)  # Flatten features
            features.append(image_features.cpu())  # Move to CPU
            labels.append(label.cpu())  # Move to CPU

    # Concatenate all features and labels and convert to numpy
    return torch.cat(features).numpy(), torch.cat(labels).numpy()
```

```python
# Paths and parameters
data_dir = './../../dataset/raw_data/train'  # Your dataset path (5 subfolder classes)
batch_size = 16
num_epochs = 10
image_size = 224  # Input size for Swin Transformer
```

```python
# Transformations: Random crop for training, Resize for validation
train_transform = transforms.Compose([
    #transforms.RandomResizedCrop(image_size),
    transforms.Resize((image_size, image_size)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])
val_transform = transforms.Compose([
    transforms.Resize((image_size, image_size)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])
```

```python
# Load the dataset
dataset = datasets.ImageFolder(root=data_dir)

# Ensure the dataset classes are in the correct order (1, 2, 3, 4, 5)
dataset.classes.sort(key=lambda x: int(x))  # Sort the class names numerically if they are numeric strings

# Create a mapping from numerical labels to class names
label_to_class_name = {idx: class_name for idx, class_name in enumerate(dataset.classes)}

# Assign transformations
dataset.transform = train_transform

# Create DataLoaders
train_data = DataLoader(dataset)
```

```python
# Assuming train_loader is your DataLoader for training data
X, labels = extract_features(train_data)
```

```python
# Load the ResNet-50 model for feature extraction
resnet_model = models.resnet50(pretrained=True)
resnet_model = torch.nn.Sequential(*(list(resnet_model.children())[:-1]))  # Remove the last layer
resnet_model.to(device)
resnet_model.eval()  # Set to evaluation mode
```

```python
from transformers import SwinForImageClassification, SwinConfig
# Define the Swin transformer model
swin_model = SwinForImageClassification.from_pretrained(
    'microsoft/swin-base-patch4-window7-224',
    num_labels=5,
    ignore_mismatched_sizes=True
)
swin_model.to(device)
swin_model.eval()
```

```python
import torch.optim as optim
import torch.nn as nn

# Define the optimizer and loss function for both models
learning_rate = 1e-5  # Adjust as necessary
weight_decay = 1e-4

# Optimizers for Swin and ViT models
swin_optimizer = optim.Adam(swin_model.parameters(), lr=learning_rate, weight_decay=weight_decay)
vit_optimizer = optim.Adam(vit_model.parameters(), lr=learning_rate, weight_decay=weight_decay)

class_counts = [277,324,581,703,291]

total_samples = sum(class_counts)
class_weights = [total_samples / (len(class_counts) * count) for count in class_counts]
class_weights = torch.FloatTensor(class_weights).to(device)
# Loss function
criterion = nn.CrossEntropyLoss(weight=class_weights)
```

```python

def train_model(model, optimizer, criterion, train_loader, val_loader, num_epochs, patience=5):
    best_val_loss = float('inf')  # Initialize best validation loss
    patience_counter = 0
    
    model.train()
    
    # Training loop
    for epoch in range(num_epochs):
        running_loss = 0.0
        correct = 0
        total = 0
        
        # Training phase
        for images, labels in train_loader:
            images, labels = images.to(device), labels.to(device)

            optimizer.zero_grad()
            outputs = model(images).logits
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()

            running_loss += loss.item()
            _, predicted = torch.max(outputs, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()

        train_loss = running_loss / len(train_loader)
        train_acc = 100 * correct / total
        train_loss_history.append(train_loss)
        train_acc_history.append(train_acc)

        # Validation phase
        model.eval()
        val_loss = 0.0
        val_correct = 0
        val_total = 0
        all_preds = []
        all_labels = []
        
        with torch.no_grad():
            for images, labels in val_loader:
                images, labels = images.to(device), labels.to(device)
                outputs = model(images).logits
                loss = criterion(outputs, labels)
                val_loss += loss.item()
                
                _, predicted = torch.max(outputs, 1)
                val_total += labels.size(0)
                val_correct += (predicted == labels).sum().item()

                all_preds.extend(predicted.cpu().numpy())
                all_labels.extend(labels.cpu().numpy())

        val_loss /= len(val_loader)
        val_acc = 100 * val_correct / val_total
        val_loss_history.append(val_loss)
        val_acc_history.append(val_acc)

        # Print metrics for this epoch
        print(f'Epoch {epoch + 1}/{num_epochs}, '
              f'Train Loss: {train_loss:.4f}, Train Acc: {train_acc:.2f}%, '
              f'Val Loss: {val_loss:.4f}, Val Acc: {val_acc:.2f}%')

        # Check for early stopping
        if val_loss < best_val_loss:
            best_val_loss = val_loss
            patience_counter = 0  # Reset counter
            # Optionally, save the best model here
            # torch.save(model.state_dict(), 'best_model.pth')
        else:
            patience_counter += 1
        
        if patience_counter >= patience:
            print(f'Early stopping triggered after {epoch + 1} epochs.')
            break

        # Confusion matrix for validation set
        cm = confusion_matrix(all_labels, all_preds)
        plt.figure(figsize=(8, 6))
        sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=dataset.classes, yticklabels=dataset.classes)
        plt.xlabel('Predicted')
        plt.ylabel('True')
        plt.title(f'Confusion Matrix (Epoch {epoch + 1})')
        plt.show()

# Call the function with the appropriate parameters
# train_model(model, optimizer, criterion, train_loader, val_loader, num_epochs)
```

```python
def get_ensemble_predictions(loaders):
    all_preds = []

    for loader in loaders:
        for images, _ in loader:
            images = images.to(device)  # Ensure images are on the correct device

            # Predictions from ResNet-50
            with torch.no_grad():
                resnet_features = resnet_model(images).view(images.size(0), -1)
                resnet_preds = svm_model.predict(resnet_features.detach().cpu().numpy())
            
            # Predictions from Swin Transformer
            with torch.no_grad():
                swin_outputs = swin_model(images).logits
                swin_preds = torch.argmax(swin_outputs, dim=1).cpu().numpy()

            # Predictions from ViT
            with torch.no_grad():
                vit_outputs = vit_model(images).logits
                vit_preds = torch.argmax(vit_outputs, dim=1).cpu().numpy()

            # Collect predictions for this batch
            all_preds.append((resnet_preds, swin_preds, vit_preds))
    
    # Flatten the predictions
    resnet_preds_flat = np.concatenate([preds[0] for preds in all_preds])
    swin_preds_flat = np.concatenate([preds[1] for preds in all_preds])
    vit_preds_flat = np.concatenate([preds[2] for preds in all_preds])

    # Average the predictions
    final_preds = np.round((resnet_preds_flat + swin_preds_flat + vit_preds_flat) / 3).astype(int)

    return final_preds
```

```python
from sklearn.metrics import accuracy_score

# Assuming y_test contains the true labels for the test set
accuracy = accuracy_score(y_test, ensemble_predictions)
print(f'Ensemble Model Accuracy: {accuracy:.2f}')
```
